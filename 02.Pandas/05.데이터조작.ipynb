```python
np.random.seed = 2023
df = pd.DataFrame(np.random.randint(40, 101, size=(4, 5)),
                index=list('abcd'), columns=list('VWXYZ'))
df
```

- 행/열 합계

```python
df
```

```python
# 행 방향(열의 합계)
df.sum(axis=0)
```

```python
# 열 방향(행의 합계)
df.sum(axis=1)
```

```python
# 총점 열 만들기
df['총점'] = df.sum(axis=1)
df
```

```python
# 평균 열 만들기
df['평균'] = df[['V', 'W', 'X', 'Y', 'Z']].mean(axis=1)
df
```

```python
df['평균2'] = df.iloc[:, :5].mean(axis=1)
df
```

```python
# 행 추가 방법
df.loc['합계', :] = df.sum()
df
```

```python
# 타이타닉호의 승객 중 나이 20살을 기준으로 성인(adult)과 미성년자(child)를 구별하는 라벨 열
titanic['adult/child'] = titanic.age.apply(lambda x: 'adult' if x >= 20 else 'child')
titanic.tail()
```

```python
data = {
    'price': ['1,200', '2,500', '3,300'],
    'text': ['03.입출력.ipynb', '04.고급인덱싱.ipynb', '05.데이터조작.ipynb']
}
df3 = pd.DataFrame(data)
df3
```

```python
int('1,200'.replace(',',''))
```

```python
df3['정수'] = df3.price.apply(lambda x: int(x.replace(',','')))
df3
```

```python
df3['구둣점제거'] = df3.text.apply(lambda x: x.replace('.',''))
df3
```

- 연습 문제 4.4.4
    - 타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category1 열을 만들어라. category1 카테고리는 다음과 같이 정의된다.
    - 20살이 넘으면 성별을 그대로 사용한다.
    - 20살 미만이면 성별에 관계없이 “child”라고 한다.

```python
del titanic['adult/child']
titanic.head()
```

```python
titanic.sex
```

```python
titanic['category1'] = titanic.apply(lambda x: x.sex if x.age >= 20 else 'child', axis=1)
titanic.head()
```
