- 연습 문제 4.4.4
    - 타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category1 열을 만들어라. category1 카테고리는 다음과 같이 정의된다.
    - 20살이 넘으면 성별을 그대로 사용한다.
    - 20살 미만이면 성별에 관계없이 “child”라고 한다.

```python
del titanic['adult/child']
titanic.head()
```

```python
titanic.sex
```

```python
titanic['category1'] = titanic.apply(lambda x: x.sex if x.age >= 20 else 'child', axis=1)
titanic.head()
```

```python
category2 = []
for i in titanic.index:
    if titanic.age[i] >= 20:
        category2.append(titanic.sex[i])
    else:
        category2.append('child')
titanic['category2'] = category2
titanic.tail()
```

- 실수 값을 카테고리 값으로 변환

```python
ages = [0, 2, 10, 21, 23, 37, 31, 61, 20, 41, 32, 101]
```

```python
bins = [1, 20, 30, 50, 70, 100]                     # 1 < 미성년자 <= 20
labels = ["미성년자", "청년", "중년", "장년", "노년"]
cats = pd.cut(ages, bins, labels=labels)
cats
```

```python
df5 = pd.DataFrame(ages, columns=["ages"])
df5["age_cat"] = pd.cut(df5.ages, bins, labels=labels)
df5
```

```python
df5['one_over_n'] = pd.qcut(ages, 4, labels=['Q1','Q2','Q3','Q4'])
df5
```
