- 경기도 인구 소멸 비율(2022. 12 기준)
    - 가임기 여성(만 20~39세) 인구수 / 60세 이상 노년 인구
    - 경기도의 인구 소멸 위기지역 Top 10
    - 위기지역: 기초자치단체 (시, 군)
    - 특례시: 수원, 고양, 용인 (구)
    - 행정구: 성남, 안산, 안양 (구)

```python
d_pct = pd.read_csv('data/주민등록인구집계현황.csv', encoding='euc-kr')
d_pct.head(3)
```

```python
d_pct['인구 소멸 비율'] = (d_pct.loc[:, '20~29세 (여)':'30~39세 (여)'].sum(axis=1) / d_pct.loc[:, '60~69세':'100세 이상'].sum(axis=1)).round(2)
d_pct = d_pct[(d_pct['행정구역구분명'] != '도') & (d_pct['행정구역구분명'] != '읍면동')]
d_pct.head(3)
```

```python
d_pct = d_pct[['행정구역명', '인구 소멸 비율']]
d_pct
```
