1. cctv 데이터를 다음과 같이 만들어 cctv.csv 파일에 저장하세요.
    - 구별    CCTV댓수  최근증가율
    - 1 종로구 1715         20.69
    - 2 중구    2447         78.98
    - 3 용산구 2611          78.64
        - 첫번째 행인 계는 삭제한다.
        - 최근증가율은 최근 3개년간의 CCTV 댓수의 합을 그 이전 모든 년도의 CCTV 댓수의 합으로 나눈 값으로 백분율 값이다.
        - 구별은 기존의 구분 열을 CCTV댓수는 기존의 총계 열을 이름만 변경한 것이다.

​

```python
cctv = pd.read_csv('data/서울시 자치구 년도별 CCTV 설치 현황_221231기준.csv', encoding='euc-kr')
cctv.head()
```

```python
# cctv.drop([0], inplace=True)
cctv = cctv[cctv.구분 != '계']
cctv.head()
```

```python
columns = '구분,총계,2013년 이전 설치된 CCTV,2013년,2014년,2015년,2016년,2017년,2018년,2019년,2020년,2021년,2022년'.split(',')
for col in columns[1:]:
    cctv[col] = cctv[col].str.replace(',','').astype(int)
cctv.head()
```

```python
# cctv.loc[:, '2020년':'2022년'].sum(axis=1)
# (cctv.iloc[:, 10:].sum(axis=1) / cctv.iloc[:, 2:10].sum(axis=1) * 100).round(2)
cctv['최근증가율'] = (cctv.T[10:13].sum() / cctv.T[2:10].sum() * 100).apply(lambda x: round(x, 2))
cctv.head()
```

```python
# cctv = cctv[['구분', '총계', '최근증가율']]
cctv.drop(columns=['2013년 이전 설치된 CCTV','2013년','2014년','2015년','2016년','2017년','2018년','2019년','2020년','2021년','2022년'], axis=1, inplace=True)
cctv.head()
```

```python
# cctv.rename(columns={'구분':'구별', '총계':'CCTV댓수'}, inplace=True)
cctv.columns = ['구별', 'CCTV댓수', '최근증가율']
cctv.head()
```

```python
cctv['구별'] = cctv['구별'].str.replace(' ','')
cctv.head()
```

```python
cctv.to_csv('data/cctv.csv')
cctv = pd.read_csv('data/cctv.csv')
del cctv['Unnamed: 0']
cctv.head()
```

2. 서울시 인구 데이터를 다음과 같이 만들어 pop.csv 파일에 저장하세요.
  - 구별     인구수  내국인  외국인  고령자  외국인비율  고령자비율
  - 종로구 153789 144683  9106    27818   5.92           18.09
  - 중구    131787 122499  9288    24392   7.05           18.51
  - 용산구 237285 222953  14332   39070   6.04           16.47
    - 첫번째 행인 합계는 삭제한다.
    - 외국인비율은 외국인 수를 총 인구수로 나눈 백분율 값이다.
    - 고령자비율은 65세 이상 인구 수를 총 인구수로 나눈 백분율 값이다.

```python
# pop = pd.read_csv('data/서울시인구.tsv', sep='\t', skiprows=2) -> 위의 두 행 스킵
df = pd.read_csv('data/서울시인구.tsv', sep='\s+')
df.head()
```

```python
# 결측치 확인
# df.isna.sum().sum()
```

```python
# pop = pop[['자치구', '계', '계.1', '계.2', '65세이상고령자']]
df.drop(['세대당인구', '기간', '세대', '인구.1', '인구.2', '인구.4', '인구.5', '인구.7', '인구.8'], axis=1, inplace=True)
df.head()
```

```python
# pop.drop([0], inplace=True)
df = df[(df['자치구'] != '자치구') & (df['자치구'] != '합계')]
df.head()
```

```python
df.columns = ['구별', '인구수', '내국인', '외국인', '고령자']
df.head()
```

```python
# for column in pop.columns[1:]:
#     pop[column] = pop[column].str.replace(',','').astype(int)
columns = '구별,인구수,내국인,외국인,고령자'.split(',')
for col in columns[1:]:
    df[col] = df[col].str.replace(',','').astype(int)
df.head()
```

```python
df['외국인비율'] = (df.외국인 / df.인구수 * 100).round(2)
df.head()
```

```python
df['고령자비율'] = (df.고령자 / df.인구수 * 100).round(2)
df.head()
```

```python
# df.to_csv('data/pop.csv', index=False)
# df = pd.read_csv('data/pop.csv')
df.to_csv('data/pop.csv')
pop = pd.read_csv('data/pop.csv')
del pop['Unnamed: 0']
pop.head()
```
