### YouTube Ranking 시각화

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('data/20230425_Youtube랭킹.csv')
df.head(3)
```

- 구독자수 Top 20 - barh 차트로

```python
plt.barh(df.head(20).channel, df.head(20).구독자수)
plt.title('구독자수 Top 20 channel')
plt.show()
```

```python
dfs = df.head(20).sort_values(by='구독자수')
plt.barh(dfs.channel, dfs.구독자수)
plt.title('구독자수 Top 20 channel')
plt.show()
```

```python
sns.barplot(y='channel', x='구독자수', data=df.head(20))
plt.title('구독자수별 Top 20');
```

```python
df1 = df.sort_values(by='view', ascending=False)
```

```python
sns.barplot(y='channel', x='view', data=df1.head(20))
plt.title('조회수별 Top 20');
```

```python
df2 = df.sort_values(by='video', ascending=False)
```

```python
sns.barplot(y='channel', x='video', data=df2.head(20))
plt.title('비디오수별 Top 20');
```

```python
d_cc = df.category.value_counts().head(10).to_frame()
d_cc
```

```python
sns.barplot(y=d_cc.index, x='count', data=d_cc)
plt.title('카테고리별 채널수 Top 10');
```

```python
d_cs = df.groupby('category')['구독자수'].sum().sort_values(ascending=False).head(10).to_frame()
d_cs
```

```python
sns.barplot(y=d_cs.index, x=d_cs.구독자수)
plt.title('카테고리별 구독자수 Top 10');
```
