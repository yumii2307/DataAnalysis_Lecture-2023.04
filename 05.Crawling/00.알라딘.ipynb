```python
def conver(s):
    if '지은이' in s:
        s = s.split('(지은이)')[0].strip()
    else:
        s = s.split('(엮은이)')[0].strip()
    return s
```

```python
lines = []
for index, li in enumerate(lis):
    try:
        title = li.select_one('.ss_book_list > ul > li > a > b').get_text()
        author = conver(li.select_one('.ss_book_list > ul > li:nth-child(3)').get_text())
        company = li.select_one('.ss_book_list > ul > li:nth-child(3)').get_text().split('|')[-2].strip()
        payment = li.select_one('.ss_book_list > ul > li:nth-child(4) > span').get_text().replace(',','')
        dc_payment = li.select_one('.ss_book_list > ul > li > .ss_p2 > b > span').get_text().replace(',','')
        mileage = li.select_one('.ss_book_list > ul > li > .ss_p:nth-child(4)').get_text()
        lines.append({'제목': title, '작가': author, '출판사': company,
                        '가격': int(payment), '판매가격': int(dc_payment), '마일리지': int(mileage)})
    except:
        continue
a = pd.DataFrame(lines)
a
```

```python
lines = []
base_url = 'https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0'
for pg in tqdm(range(1, 21)):
    url = f'{base_url}&page={pg}&cnt=1000&SortOrder=1'
    result = requests.get(url)
    soup = BeautifulSoup(result.text, 'html.parser')
    lis = soup.select('.ss_book_box')
    li = lis[0]
    for li in lis:
        try:
            title = li.select_one('.ss_book_list > ul > li > a > b').get_text()
            author = conver(li.select_one('.ss_book_list > ul > li:nth-child(3)').get_text())
            company = li.select_one('.ss_book_list > ul > li:nth-child(3)').get_text().split('|')[-2].strip()
            payment = li.select_one('.ss_book_list > ul > li:nth-child(4) > span').get_text().replace(',','')
            dc_payment = li.select_one('.ss_book_list > ul > li > .ss_p2 > b > span').get_text().replace(',','')
            mileage = dc_payment = li.select_one('.ss_book_list > ul > li > .ss_p:nth-child(4)').get_text()
            lines.append({'제목': title, '작가': author, '출판사': company,
                            '가격': int(payment), '판매가격': int(dc_payment), '마일리지': int(mileage)})
        except:
            continue
aladin = pd.DataFrame(lines)
aladin.head(50)
```
